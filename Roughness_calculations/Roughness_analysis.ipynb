# Analysis of roughness data
This script takes the output from QGIS and compares them to in situ measurements.

```python
from io import StringIO
import pandas as pd
import glob
import os
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data_folder=r'Input\Extracted_from_DEM'
chain_data_folder=r'Input'
```

The following imports 2d and 3d transect data calculated from QGIS (exported in csv for 2d and html for 3d).

```python
dirname = os.path.realpath('')+data_folder
all_files = glob.glob(data_folder + "\*.html")
roughness_data = []

for filename in all_files:
 with open(filename, 'r') as f:
     html_string = f.read()
    
 html_string=html_string.replace('<html><head><meta http-equiv="Content-Type" content="text/html; charset=utf-8" /></head><body><p>','')
 html_string=html_string.replace('</br>','')
 html_string=html_string.replace('</p></body></html>','')
 html_string=html_string.replace('\n\t','')

 data = StringIO(html_string)
 df = pd.read_csv(data, sep=":",header=None).drop([2])
 df.columns=['Description','Area_sqm']
 df['Area_sqm']=df['Area_sqm'].astype(float)
    
 sitename = str(os.path.basename(filename)).replace('.html','')
 area2D=df.Area_sqm[1]-df.Area_sqm[0]
 area3D=df.Area_sqm[4]
 roughness_data.append([sitename,area2D,area3D])

Qgis_out_3d = pd.DataFrame(roughness_data, columns=['site_name','2d_area_sqm','3d_area_sqm'])
```

```python
dirname = os.path.realpath('')+data_folder
all_files = glob.glob(data_folder + "\*.csv")
roughness_data = []

for filename in all_files:
 with open(filename, 'r') as f:
    csv = pd.read_csv(filename,)
 sitename = str(os.path.basename(filename)).replace('.csv','')
 map_legth=max(csv.DIST)
 real_length=max(csv.DIST_SURF)
 roughness_data.append([sitename,map_legth,real_length])

Qgis_out_2d = pd.DataFrame(roughness_data, columns=['site_name','Map_length_m','Real_length_m'])
```

```python
chain=pd.read_excel(chain_data_folder+'\Chain_length.xlsx')
```

The datasets are then joined and a csv file with all the data is exported.

```python
joined=chain.join(Qgis_out_2d.set_index('site_name'), on='Transect').join(Qgis_out_3d.set_index('site_name'), on='Transect')
joined['chain_roughness']=joined['Chain_length']/joined['Map_length_m']
joined['drone_transect_roughness']=joined['Real_length_m']/joined['Map_length_m']
joined['drone_surface_roughness']=joined['3d_area_sqm']/joined['2d_area_sqm']
export=joined.rename(columns={"Transect": "Site name",
                              "Chain_length": "Chain length [m]",
                              "Map_length_m":"Map length [m]",
                              "Real_length_m":"DSM length [m]",
                              "2d_area_sqm":"Flat surface area [sqm]",
                              "3d_area_sqm":"Real surface area [sqm]",
                              "chain_roughness":"Chain roughness",
                              "drone_transect_roughness":"Drone transect roughness",
                              "drone_surface_roughness":"Drone surface roughness"})
export.to_csv('Roughness_total.csv')
export
```

Chain and drone roughnesses are compared in two graphs.

```python
chain_roughness=pd.DataFrame()
chain_roughness['roughness']=joined['chain_roughness']
chain_roughness['Transect']=joined['Transect']
chain_roughness['type']='Chain roughness'

drone_transect_roughness=pd.DataFrame()
drone_transect_roughness['roughness']=joined['drone_transect_roughness']
drone_transect_roughness['Transect']=joined['Transect']
drone_transect_roughness['type']='Drone 2d roughness'

drone_surface_roughness=pd.DataFrame()
drone_surface_roughness['roughness']=joined['drone_surface_roughness']
drone_surface_roughness['Transect']=joined['Transect']
drone_surface_roughness['type']='Drone 3d roughness'

bar_df=chain_roughness.append(drone_transect_roughness).append(drone_surface_roughness)

ax = sns.barplot(x='Transect', y='roughness', hue='type', data=bar_df,linewidth=1,edgecolor=(0,0,0))
#ax = sns.lineplot(x='Transect', y='roughness', hue='type', data=bar_df,markers=True,linewidth=1.5)

plt.setp(ax.get_xticklabels(), rotation=90)
ax.legend_.set_title(None)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Site')
plt.ylabel('Roughness')
plt.savefig('roughness_comparison.svg')
```

```python
ax=sns.regplot(x=chain_roughness['roughness'], y=drone_transect_roughness['roughness'],scatter=True,ci=95,label='2d roughness')
ax1=sns.regplot(x=chain_roughness['roughness'], y=drone_surface_roughness['roughness'],scatter=True,ci=95,label='3d roughness')

#Thanks https://stackoverflow.com/questions/25497402/adding-y-x-to-a-matplotlib-scatter-plot-if-i-havent-kept-track-of-all-the-data
lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),
    np.max([ax.get_xlim(), ax.get_ylim()]),
]
ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
ax.set_aspect('equal')
ax.set_xlim(lims)
ax.set_ylim(lims)

plt.xlabel('Chain and tape roughness')
plt.ylabel('DSM-derived roughness')
plt.legend()
plt.savefig('comparison.svg')
```
